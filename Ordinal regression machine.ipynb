```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('ordinaldata.txt', 'r') as f :
    str_d = f.read()
    sep_data = [item.strip('[]').split() for item in str_d.split('\n')]
    res = []
    for item in sep_data : 
        item = [float(x) for x in item]
        res.append(item)
res = np.array(res)
```

```python
plt.figure()
plt.scatter(res[:, 0],res[:, 1], c = res[:, -1])
plt.show()
```

```python
class OrdinalRegressionMachine : 
    def __init__(this,X, y):
        this.X = X
        this.y = y
        
    def printclass(this):
        print('class works fine')
        
        
```

```python
obj = OrdinalRegressionMachine(res[:, 0:2],res[:, -1])
obj.printclass()
```
